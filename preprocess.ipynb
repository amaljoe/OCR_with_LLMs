```python
import easyocr
import cv2

import pytesseract
from pytesseract import Output

import matplotlib.pyplot as plt
```

```python
reader = easyocr.Reader(['en'])
results = reader.readtext("data/handwritten.png")

print(results)

for bbox, text, confidence in results:
    print(f"Detected : {text} ({confidence:.2f})    Coordinates: {bbox}")

image = cv2.imread("data/handwritten.png")
paragraph1 = ""
# Original vs Detected
for bbox, text, confidence in results:
    top_left = tuple(map(int, bbox[0]))
    bottom_right = tuple(map(int, bbox[2]))
    image = cv2.rectangle(image, top_left, bottom_right, (0, 255, 0), 2)
    image = cv2.putText(image, text, top_left, cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 255), 2, cv2.LINE_AA)
    paragraph1 += text + " "

plt.figure(figsize=(10, 10))
plt.imshow(image)
```

```python
reader = easyocr.Reader(['en'])
results = reader.readtext("data/paragraph.png")

for bbox, text, confidence in results:
    print(f"Detected : {text} ({confidence:.2f})    Coordinates: {bbox}")

image = cv2.imread("data/paragraph.png")

# Original vs Detected
for bbox, text, confidence in results:
    top_left = tuple(map(int, bbox[0]))
    bottom_right = tuple(map(int, bbox[2]))
    image = cv2.rectangle(image, top_left, bottom_right, (0, 255, 0), 2)
    image = cv2.putText(image, text, top_left, cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 255), 2, cv2.LINE_AA)


plt.figure(figsize=(10, 10))
plt.imshow(image)
```

```python
image = cv2.imread("data/handwritten.png")
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
results = pytesseract.image_to_data(image, output_type=Output.DICT)
print(results['text'])

paragraph2 = ""
for i in range(len(results["text"])):
    text = results["text"][i]
    confidence = float(results["conf"][i])
    if text.strip() and confidence > 0:
        x, y, w, h = results["left"][i], results["top"][i], results["width"][i], results["height"][i]
        image = cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)
        image = cv2.putText(image, text, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)
        paragraph2 += text + " "

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 10))
plt.imshow(image_rgb)
plt.show()
```
