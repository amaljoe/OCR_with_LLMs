```python
from PIL import Image

image = Image.open('data/iam_form.png')
image
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

textLines = list(np.array(image).mean(axis=1) < 240)
heights = []
last = textLines.index(True)
for i in range(last, len(textLines) - 1):
    if textLines[i] and not textLines[i+1]:
        heights.append(i - last)
    if not textLines[i] and textLines[i+1]:
        last = i + 1

pd.Series(heights).hist(bins=range(0, 100, 2))
```
